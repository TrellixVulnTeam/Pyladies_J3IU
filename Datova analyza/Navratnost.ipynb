```python
import pandas as pd
```

```python
sreality_Q1 = pd.read_csv("data/data_pro_pr_od_2019-01-01_do_2019-03-31_pl.csv")
sreality_Q2 = pd.read_csv("data/data_pro_pr_od_2019-04-01_do_2019-06-30_pl.csv")

sreality_Q1['kvartal'] = '1/2019'
sreality_Q2['kvartal'] = '2/2019'

sreality = pd.concat([sreality_Q1, sreality_Q2])
sreality = sreality[~sreality["Id inzerátu"].duplicated()]
```

```python
sreality
```

```python
sreality['cena/metr2'] = sreality['Celková cena']//sreality['Užitná plocha']
sreality
```

```python
byty = sreality[sreality["Kategorie"] == "Byty"]
byty
```

```python
byty_filtr = byty.groupby(['Okres', 'Typ']).mean().unstack()
byty_filtr
```

```python
byty_filtr['návratnost'] = byty_filtr['cena/metr2', 'Prodej']/ (12*byty_filtr['cena/metr2', 'Pronájem'])
byty_filtr
```

```python
byty_filtr.sort_values('návratnost')
```

```python
byty_filtr = byty_filtr.drop(columns=['Unnamed: 0','Id inzerátu','Plocha pozemku'])
```
