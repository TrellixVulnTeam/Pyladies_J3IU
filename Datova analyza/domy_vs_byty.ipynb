```python
import pandas as pd
import numpy as np
```

```python
sreality_Q1 = pd.read_csv("data/data_pro_pr_od_2019-01-01_do_2019-03-31_pl.csv")
sreality_Q2 = pd.read_csv("data/data_pro_pr_od_2019-04-01_do_2019-06-30_pl.csv")

sreality_Q1['kvartal'] = '1/2019'
sreality_Q2['kvartal'] = '2/2019'

sreality = pd.concat([sreality_Q1, sreality_Q2])
sreality = sreality[~sreality["Id inzerátu"].duplicated()]
```

```python
sreality
```

```python
byty = sreality[(sreality["Kategorie"] == "Byty")&(sreality["Typ"] == "Prodej")]
domy = sreality[(sreality["Kategorie"] == "Domy")&(sreality["Typ"] == "Prodej")]
byty = byty.groupby('Kraj').agg({"Celková cena": np.mean, "Užitná plocha": np.mean, "Id inzerátu": len})
domy = domy.groupby('Kraj').agg({"Celková cena": np.mean, "Užitná plocha": np.mean, "Id inzerátu": len})

byty
domy
```
